```python
def grayDecode(n):
  p = n
  n = n >> 1
  while n != 0:
    p ^= n
    n = n >> 1
  return p
```

```python
def grayEncode(n):
  return n ^ (n >> 1)
```

```python
def graycode2Coord(g, k):
  i_ = (int(g) >> int(k / 2)) & 0xFFFFFFFF
  j_ = g ^ (i_ << int(k / 2))

  i = grayDecode(i_)
  j = grayDecode(j_)
  return (i,j)
```
